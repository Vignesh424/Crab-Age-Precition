```python
#import Packages
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import *
import pickle
```

```python
#Read the Data
data = pd.read_csv('/content/CrabAgePrediction.csv')
```

```python
#Split it into Features and Target
X = data.drop(columns=['Age','Sex'],axis=1)
Y = data['Age']
```

```python
#Split into Train and Test Datasets
X_train , X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=2)
```

```python
#Use Regression Model
rfr = RandomForestRegressor()
rfr.fit(X,Y)
```

```python
#Predictor
Y_pred = rfr.predict(X_test)
```

```python
#Check the Score
rfr.score(X_test, Y_test)
```

```python
print("Mean Square Erorr:", metrics.mean_squared_error(Y_test, Y_pred))
print("Root Mean Square Erorr:", np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))
print("Mean Absolute Error:", metrics.mean_absolute_error(Y_test, Y_pred))
```
